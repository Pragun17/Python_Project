# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
df = pd.read_csv('summer.csv', usecols= ['City'])
new_df = df.drop_duplicates(subset = ['City'])
print(new_df)
cities_num = len(new_df)
print("The number of cities in which summer olympics is held is: " + str(cities_num))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = pd.read_csv('summer.csv')
gold_md = df[df['Medal']=='Gold']
data = []
for sport in gold_md['Sport'].unique():
  data.append([sport, len(gold_md[gold_md['Sport']==sport])])
data = pd.DataFrame(data, columns=['Sport', 'Gold']).sort_values(by='Gold', ascending=False).head()
data.plot(x='Sport', y='Gold', kind='bar', figsize=(10, 5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending=False).head().plot.bar(figsize=(10, 5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
df.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head().plot.bar(figsize=(10, 5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
gold_md.groupby('Athlete').count()['Medal'].sort_values(ascending=False).head(5).plot.bar(figsize=(10, 5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
gold_md[gold_md['Country']=='IND']['Year'].min()
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
df.groupby("Event").count()['Athlete'].sort_values(ascending=False).head(5).plot.bar(figsize=(10, 5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
Female_gold = gold_md[gold_md['Gender']=='Women']
data=[]
for sport in Female_gold['Sport'].unique():
  data.append([sport, len(Female_gold[Female_gold['Sport']==sport])])

data= pd.DataFrame(data, columns=['Sport', 'Female_gold']).sort_values(by='Female_gold', ascending=False).head()
data.plot(x='Sport' ,y='Female_gold', kind='bar', figsize=(10, 5))
```
